```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from highlight_text import fig_text
```

```python
mpl.rcParams['figure.dpi']= 200
```

```python
palette = {
    'Comparison' : '#f03e3e',
    'Shipman' : '#1c7ed6'
}
```

```python
df = pd.read_csv('00-2-shipman-times-x.csv')
```

```python
df_melted = df.melt(id_vars='Hour', var_name='Doctor', value_name='Percent')
```

```python
fig, ax = plt.subplots()

sns.lineplot(data=df_melted, x='Hour', y='Percent', hue='Doctor', palette=palette, lw=2, ax=ax, legend=False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plot_title = 'Percentage of deaths by hour of day\n<Shipman> and <comparison GPs>'

highlight_textprops = [
    {"color": palette["Shipman"], "fontweight": "bold"},
    {"color": palette["Comparison"], "fontweight": "bold"}
    ]

fig_text(x=0.125, y=0.9, s=plot_title, highlight_textprops=highlight_textprops, fontsize=12, va='bottom');
```
