```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
mpl.rcParams['figure.dpi']= 200
```

```python
df = pd.read_csv('01-1-child-heart-survival-x.csv')
```

```python
df['Percentage'] = df['Operations'] / df['Operations'].sum()
```

```python
df = df.sort_values('Percentage', ascending=False)
```

```python
fig, ax = plt.subplots()

sns.barplot(data=df, y='Hospital', x='Percentage', ax=ax, palette='RdPu_r')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(
    axis='y',
    which='both',
    left=False,
    labelleft=True)

ax.set_ylabel(None)

ax.xaxis.set_major_formatter(mtick.PercentFormatter(1,0))

fig.suptitle('Percentage of all operations in 2012-15 that are carried out in each hospital', x=0, ha='left');
```
