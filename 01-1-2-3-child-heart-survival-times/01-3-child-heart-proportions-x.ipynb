```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
mpl.rcParams['figure.dpi']= 200
```

```python
df = pd.read_csv('01-1-child-heart-survival-x.csv')
```

```python
df['thirty_day_survival_pct'] = df['ThirtyDaySurvival'] / 100
```

```python
df = df.sort_values('thirty_day_survival_pct', ascending=False)
```

```python
fig, ax = plt.subplots()

sns.barplot(data=df, y='Hospital', x='thirty_day_survival_pct', ax=ax, palette='RdPu_r')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(
    axis='y',
    which='both',
    left=False,
    labelleft=True)

ax.xaxis.set_major_formatter(mtick.PercentFormatter(1,0))

ax.set_xlabel('Thirty day survival')
ax.set_ylabel(None)

fig.suptitle('Thirty day survival rates at each hospital', x=0, ha='left');
```

```python
fig, ax = plt.subplots()

sns.scatterplot(data=df, y='Hospital', x='thirty_day_survival_pct', ax=ax, hue='thirty_day_survival_pct', palette='RdPu', edgecolor='black', legend=False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.xaxis.set_major_formatter(mtick.PercentFormatter(1,0))

ax.set_xlim(xmin=0.86, xmax=1)

ax.set_xlabel('Thirty day survival')
ax.set_ylabel(None)

fig.suptitle('Thirty day survival rates at each hospital', x=0, ha='left');
```
