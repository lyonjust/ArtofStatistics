```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from highlight_text import fig_text
```

```python
mpl.rcParams['figure.dpi']= 200
```

```python
df = pd.read_csv('02-1-bean-data-full-x.csv', header=None)
```

```python
def plot_strip_box_hist(log_scale=False):
    fig, ax = plt.subplots(nrows=3, ncols=1, sharex=True)

    sns.stripplot(data=df, x=0, ax=ax[0], alpha=0.5, size=2, color='white', edgecolor='black', linewidth=0.5, jitter=1)
    ax[1].boxplot(x=df[0], vert=False, widths=0.7, flierprops={'markersize' : 2, 'linewidth' : 0.5}, medianprops={'color' : 'black'})
    sns.histplot(data=df, x=0, ax=ax[2], color='black', log_scale=log_scale)

    if log_scale:
        ax[2].set_xscale('log')
        #ax[2].set_xticks([200, 500,2000,5000,20000])

    for axis in ax:
        axis.tick_params(
            axis='y',
            which='both',
            left=False,
            labelleft=False
            )        
        axis.set_xlabel(None)
        axis.set_ylabel(None)
```

```python
plot_strip_box_hist()
```

```python
plot_strip_box_hist(log_scale=True)
```
